```python
#Exploratory Data Analysis (EDA):
# Examinar y visualizar los datos de manera sistemática con el objetivo de descubrir patrones,
# identificar tendencias, detectar valores atípicos y extraer información relevante.

def eda_basico(df):
    print("🔍 Primeras filas del DataFrame:")
    display(df.head())
    print("📐 Dimensiones:")
    print(df.shape , "\n")
    print("🧠 Información general:")
    display(df.info())
    print("📊 Tipos de datos por columna:")
    print(df.dtypes, "\n")
    print("📉 Descripción de columnas numéricas:")
    display(df.describe())
    print("🔤 Descripción de columnas categóricas:")
    display(df.describe(include=['O']))

    print("🚫 Valores nulos por columna:")
    display(df.isnull().sum())
    print("📎 Filas duplicadas:")
    dup_count = df.duplicated().sum()
    print(f"Duplicadas: {dup_count}")
    if dup_count > 0:
        print("Ejemplo de duplicados:")
        print(df[df.duplicated()].head(), "\n")
    else:
        print("No hay filas duplicadas.\n")
```

```python
# Generamos función para quitar errorores de puntuación y cambiar a float

def pasar_a_float (columna):
    try:
        column = columna.strip().replace("Not Available", "").replace("$", "").replace(".", "").replace(",", ".").strip()
        return float(column)
    except: 
        return np.nan
```

```python
# Lista con las columnas a aplicar e iteración para aplicar a cada una. 

lista_float = ['distancia_domicilio', 'tarifa_hora', 'tarifa_diaria', 'ingreso_mensual', 'recibe_mismo_salario']

for i in lista_float:
    df[i] = df[i].apply(pasar_a_float)
```
